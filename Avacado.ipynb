# For Prediction Of Region

Separating the Data Into Features and Vectors

```python
x=AV.iloc[:,:-1]
y=AV.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.34,random_state=45)
xtrain.shape
```

# Building The Model

**MultinomialNB**

```python
from sklearn.naive_bayes import MultinomialNB
mn=MultinomialNB()
mn.fit(xtrain,ytrain)
pred=mn.predict(xtest)
print(accuracy_score(ytest,pred)*100,'%')
```

**DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
pred1=dt.predict(xtest)
print(accuracy_score(ytest,pred1)*100,'%')
```

**GaussianNB**

```python
from sklearn.naive_bayes import GaussianNB
g=GaussianNB()
g.fit(xtrain,ytrain)
pred2=g.predict(xtest)
print(accuracy_score(ytest,pred2)*100,'%')
```

**KNeighborsClassifier**

```python
from sklearn.neighbors import  KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(xtrain,ytrain)
pred3=knn.predict(xtest)
print(accuracy_score(pred3,ytest)*100,"%")
```

```python
from sklearn.svm import SVC 
```

```python
svc=SVC(kernel='rbf')
svc.fit(xtrain,ytrain)
pred4=svc.predict(xtest)
print(accuracy_score(pred4,ytest)*100,"%")
```

We are getting the best accuracy score with  DecisionTreeClassifier

Taking Decision Tree as the Final model

# For Prediction Of AveragePrice

```python
x=AV.iloc[:,1:]
y=AV.iloc[:,0]
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=45)
xtrain.shape
```

# Building the Model

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
pred=lr.predict(xtest)
from sklearn.metrics import mean_squared_error
print('Mean Squared error  is:',mean_squared_error(ytest,pred))
```

**Lasso Regression**

```python
from sklearn.linear_model import Lasso
l = Lasso()
l.fit(xtrain,ytrain)
pred2=l.predict(xtest)
print('Mean Squared error  is:',mean_squared_error(ytest,pred2))
```
